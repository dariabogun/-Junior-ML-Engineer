# Final project
## Hangman game

#### Description of the game:
For the final project, I chose the Hangman game. The principle of the game is that you should suggest a letter that can be included in this word. If such a letter is in a word, then the program writes it instead of * in the corresponding letter *, how many it occurs in the word. If there is no such letter, then a circle in a loop representing the head is added to the gallows. You keep guessing the letters until you guess the whole word. For each wrong answer, the program adds one part of the torso to the gallows (usually there are 7 of them: rope, head, torso, 2 arms and 2 legs, there is also an option with 9 parts - feet are added, as well as the longest option, when first for an unsuspected letter are drawn part of the gallows itself).

If the torso in the gallows is drawn completely, then the guessing player loses, is considered hanged: 

If the player manages to guess the word before the full body is drawn in the gallows, he wins.

#### Description of action sets:
1. The secret word should change depending on the chosen level
2. The player enters a letter
3. The player can guess the letter, but also may not guess
4. A picture of the gallows, theme, a secret word and letters that did not guess should be displayed

To run, I recommend saving the .ipynb file as .py and running it via the console with the command:
**python Final_Game_Bogun.py**

```python
import random
import os
import sys

clear = lambda: os.system('cls')

man=['''
-----------
|         
|       
|                    
|                    
|                   
|              
|              
---''','''
-----------
|         |
|        
|                   
|                   
|                 
|              
|              
---''','''
-----------
|         |
|         O
|                    
|                    
|                  
|              
|              
---
''','''
-----------
|         |
|         O
|         |           
|         |           
|                 
|              
|              
---
''','''
-----------
|         |
|         O
|        /|           
|         |           
|                  
|              
|              
---
''','''
-----------
|         |
|         O
|        /|\            
|         |           
|                 
|              
|              
---
''','''
-----------
|         |
|         O
|        /|\            
|         |           
|        /          
|              
|              
---
''','''
-----------
|         |
|         O
|        /|\            
|         |           
|        / \          
|              
|              
---
You lose
''']

ml=[]
rl=[]

def check(letter): # check function 
    global ml
    global rl
    fl1=0
    fl2=0
    # checking if a letter is a character or a number or a word
    if letter.isalpha()==True & len(letter)==1:
        # have you entered this letter before and it is in the array with the correct letters
        if letter in secret:
            for i in range(len(rl)):
                if letter==rl[i]:
                    fl1=fl1+1
            if fl1==0:
                rl.append(letter)
                return rl
            else:
                return ('Already write it')
        # have you entered this letter before and it is in the array with incorrect letters
        else:
            for i in range(len(ml)):
                if letter==ml[i]:
                    fl2=fl2+1
            if fl2==0:
                ml.append(letter)
                return ml
            else:
                return ('Already write it')
    else: 
        return ("It's not letter")

def sword(letter): # replaces gaps with guessed letters
    for i in range(len(secret)): 
        if letter==secret[i]:
            global sw
            sw = sw[:i] + letter + sw[i+1:] 
        else: continue
    return sw

def display(n,letter): # output on display
    global ml
    print('Theme: '+theme) # output of word subject (hints)
    print(man[n]) # drawing a figure of a man showing the remaining number of attempts
    print(sword(letter)) # output of a hidden word with already guessed letters, if any
    print('wrong letters: '+' '.join(ml)) # list of incorrect letters
            
def guess(): # letter input
    n=0
    while n+1!=len(man):
        letter=input('\nguess a letter: ')
        # checking if the word is already guessed
        if sword(letter)==secret:
            break
        else:
            if check(letter)=='Already write it' or check(letter)=="It's not letter":
                clear() # cleaning console
                print(check(letter))
                display(n,letter)
            else: 
                if letter in secret:
                    clear() # cleaning console
                    display(n,letter)
                else:
                    n=n+1 # attempt counter
                    clear() # cleaning console
                    display(n,letter)
    if n+1==len(man): # secret word output in case of failure
        print('The secret word was '+secret)
    else: # case of winning
        display(n,letter)
        print('Congratulations! You won!')
    start()

def level(l): # level selection function
    global wordlist
    global theme
    if l==1:
        wordlist =['sun', 'sky', 'sea', 'air', 'hill', 'tree']
        theme='Nature'
    elif l==2:
        wordlist =['apples', 'oranges', 'grapes', 'pizza', 'cheese', 'burger']
        theme='Food'
    else:
        wordlist =['understanding', 'independent', 'curious', 'obedient', 'determined', 'talkative']
        theme='Character'
    return wordlist,theme
        

def start(): # function for starting the game
    choice=int(input('\n1 - play, 2 - exit: '))
    if choice==1:
        lv=int(input('\nChoose the difficulty level (1,2 or 3): '))
        return lv
    else:
        sys.exit()

print('Hello! This is a gallows game. Try to guess the word! You have 7 attempts')
wordlist,theme = level(start())
secret = random.choice(wordlist)
sw = '*' * len(secret)
guess()
```
